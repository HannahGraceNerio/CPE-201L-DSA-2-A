```python
def traverse_array(arr):
    """Function to traverse and print all elements in the array"""
    print("Traversing the array:")
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")

def reverse(arr):
    print("Reversing all elements")
    arr.reverse()
    print(arr)

def modify(arr):
    print("Modifying all elements (multiply by 2)")
    arr[3] = [element * 2 for element in arr]
    print(arr)

def insert(arr):
    print("Inserting element 12 at index 2")
    arr.insert(2, 12)
    print(arr)

def delete(arr):
    print("Deleting element at index 2")
    if len(arr) > 2:
        del arr[2]
    print(arr)

def append(arr):
    print("Appending element 60")
    arr.append(60)
    print(arr)

def sort(arr):
    print("Sorting all elements")
    arr.sort()
    print(arr)

def extend(arr):
    print("Extending array with [70, 80, 90]")
    arr.extend([70, 80, 90])
    print(arr)

def main():
    # Initialize an array
    arr = [10, 20, 30, 40, 50]
    print("Original array:", arr)

    while True:
        print("\nMenu:")
        print("1. Traverse array")
        print("2.reverse")
        print("3.modify")
        print("4.insert")
        print("5.delete")
        print("6.append")
        print("7.sort")
        print("8.extend")
        print("9.Exit")
        choice = input("Enter your choice (1-2-3-4-5-6-7-8-9): ")

        if choice == '1':
            traverse_array(arr)
        elif choice == '2':
            reverse(arr)
        elif choice == '3':
            modify(arr)
        elif choice == '4':
            insert(arr)
        elif choice == '5':
            delete(arr)
        elif choice == '6':
            append(arr)
        elif choice == '7':
            sort(arr)
        elif choice == '8':
            extend(arr)
        elif choice == '9':
            break
        else:
          print("Invalid choice. Please enter a number between 1-2-3-4-5-6-7-8-9.")

if __name__ == "__main__":
    main()
```
